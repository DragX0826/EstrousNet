# 01 Dataset Exploration

Goal: summarize image inventory, basic intensity distributions, and cell-level annotation coverage.

```python
from pathlib import Path
import pandas as pd

raw_images = sorted(Path('../data/raw').glob('*'))
annotations_path = Path('../data/annotations/cells.csv')

print(f'Raw images: {len([p for p in raw_images if p.is_file()])}')
if annotations_path.exists():
    ann = pd.read_csv(annotations_path)
    print('Annotated cells:', len(ann))
    print(ann['cell_type'].value_counts())
else:
    print('No annotations found yet.')
```
