# 02 Segmentation Tuning

Goal: tune watershed parameters (`min_area`, `max_area`, `min_distance`) and inspect overlays.

```python
import matplotlib.pyplot as plt
from pathlib import Path
from src.utils.io import read_image
from src.preprocessing.image_cleaning import preprocess_image
from src.segmentation.cell_segmentation import segment_cells

sample = sorted(Path('../data/raw').glob('*.jpg'))[0]
img = read_image(sample)
pre = preprocess_image(img)
labels, _ = segment_cells(pre, min_area=80, max_area=20000, min_distance=10)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title('Preprocessed')
plt.imshow(pre[..., ::-1])
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Segmentation labels')
plt.imshow(labels, cmap='nipy_spectral')
plt.axis('off')
plt.tight_layout()
```
